## 3.Standardization of numerical variables

## Concept of Standardization
Z-score normalization (also called standardization) is a technique used to scale data so that it has a mean of 0 and a standard deviation of 1.This process involves subtracting the mean from each value and then dividing by the standard deviation.  It transforms the data into a standard scale, allowing comparisons between different datasets or features.

## Before Standardization

```python
"Summary Statistics Before Standardization:"
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']
numerical_columns
```

```python
scaler=StandardScaler()
scaler
```

```python
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
df[numerical_columns] 
```

## After Standardization

```python
"Summary Statistics After Standardization:"
df.describe()
```

```python
#comparison of data distributions
for column in numerical_columns:
    plt.figure(figsize=(12,6))
    plt.subplot(1,2, 1)
    plt.hist(df[column], bins=12, color='pink', edgecolor='black')
    plt.title(f'Distribution of {column} (Before Standardization)')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.subplot(1,2, 2)
    plt.hist(df[column], bins=12, color='blue', edgecolor='black')
    plt.title(f'Distribution of {column} (After Standardization)')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

## 4.conversion of categorical data into dummy variables

## Need for Converting Categorical Data into Dummy Variables (One-Hot Encoding)
Converting categorical data into dummy variables (also called one-hot encoding) is important because many machine learning algorithms and models only work with numerical data. 
In many machine learning algorithms, you can't use categorical variables directly as inputs. One-hot encoding is a method that turns the categories into a numerical format. It creates new columns for each category, using 0s and 1s to indicate whether a category is present. This way, the data is ready for machine learning models to understand and work with effectively.

```python
"Dataset Before One-Hot Encoding:"
df.head()
```

```python
df_encoded = pd.get_dummies(df)
```

```python
"Dataset After One-Hot Encoding:"
df_encoded.head()
```
